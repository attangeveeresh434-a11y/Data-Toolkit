-Here's how broadcasting works based on a set of rules:
- Padding with Ones (Rule 1):
If two arrays have a different number of dimensions, the shape of the array with fewer dimensions is padded with ones on its leading (left) side until both arrays have the same number of dimensions.
-Stretching Dimensions (Rule 2):
When comparing the shapes of the two arrays (from right to left, dimension by dimension):
If the dimensions are equal, they are compatible.

6.what are the differences  between NumPhy arrays and python lists?
- NumPy arrays and Python lists are both used to store collections of data, but they differ significantly in their characteristics and intended use cases
- 1. Data Type Homogeneity:

  - NumPy Arrays:
Store elements of a single, uniform data type (e.g., all integers, all floats). This homogeneity enables efficient storage and
 operations.

  - Python Lists:
Can store elements of different data types within the same list (e.g., an integer, a string, and a float in one list).
 2. Performance and Efficiency:

  - NumPy Arrays:Offer significantly better performance for numerical computations, especially on large datasets. This is due to their fixed size, contiguous memory storage, and optimized C implementations of operations.

  -  Python Lists:
Less efficient for numerical operations on large datasets as they are not optimized for this purpose and store elements in a less contiguous manner.
3. Memory Usage:
  - NumPy Arrays:
More memory-efficient than Python lists, particularly for large arrays, because they store homogeneous data types in contiguous memory blocks, reducing overhead.
 - Python Lists:
Can be less memory-efficient due to the overhead associated with storing heterogeneous data types and the potential for scattered memory allocation.

8.what does the term"vectorized operation" mean in NumPhy?

17. explain the  difference between apply() and map() in pandas?
- In Pandas, both map() and apply() are used to apply functions or transformations to data, but they differ in their scope and typical use cases:
 1. map():

- Scope: Exclusively used on Series objects (e.g., a single column of a  DataFrame).
- Purpose: Primarily for element-wise transformations or substitutions. It's highly efficient when you need to substitute values in a Series based on a dictionary, another Series, or a simple function.
- Input: Can accept a dictionary, a Series, or a callable (function).
- Behavior: Substitutes each value in the Series with the corresponding mapped value. If a value is not found in the mapping, it typically results in NaN.

- apply():

- Scope: Can be used on both Series and DataFrame objects.
- Purpose: More versatile than map(). It's used to apply a function along an axis (rows or columns) of a DataFrame or to each element of a Series. It's suitable for more complex operations, including aggregations or functions that require access to multiple elements or rows/columns.
Input: Accepts a callable (function).
- Behavior:On a Series: Applies the function element-wise, similar to map(), but offers more flexibility in passing arguments to the function.

- On a DataFrame: Applies the function row-wise (axis=0 by default) or column-wise (axis=1). The function receives a Series (representing a row or column) as input.

21. why is NumPhy's array slicing faster than pythons list slicing?
- Contiguous Memory Allocation:
NumPy arrays store elements of the same data type in contiguous blocks of memory. This allows for efficient access and manipulation of data, as the system can quickly locate elements within a slice without having to jump around in memory.
- Homogeneous Data Types:
NumPy arrays enforce a single, fixed data type for all elements within an array. This homogeneity allows NumPy to optimize operations, including slicing, by avoiding the need for type checking during processing
- C-level Implementation and Vectorization:
NumPy's core operations, including slicing, are implemented in highly optimized C code. This allows for much faster execution compared to Python's interpreted operations.
